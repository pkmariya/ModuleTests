```python
import pandas as pd
```

### Max Price by Category

```python
# Max Price by Category
data_df.groupby('Category')['Price'].max()
```

```python
data_df.groupby('Category')['Price'].max().reset_index()
```

```python
max_result = data_df.groupby('Category')['Price'].max().reset_index().sort_values(by='Price', ascending=True)
max_result.columns = ['Category', 'Max_Price']
print(max_result)
```

```python
max_result
```

### Average Price by Category

```python
# Average Price by Category
avg_result = data_df.groupby('Category')['Price'].mean().reset_index().sort_values(by='Price')
avg_result.columns = ['Category', 'Average']
print(avg_result)
```

```python
avg_result
```

### Min Price by Category

```python
# Min Price by Category
min_result = data_df.groupby('Category')['Price'].min().reset_index().sort_values(by='Price')
min_result.columns = ['Category', 'Min_Price']
print(min_result)
```

```python
min_result
```

### All at once - Max Price, Average and Min Price

```python
data_df.groupby('Category').agg({'Price' : ['max', 'mean', 'min']}).reset_index()
```

```python
result = data_df.groupby('Category').agg({'Price' : ['max', 'mean', 'min']}).reset_index()
result.columns = ['Category', 'Max_Price', 'Average', 'Min_Price']
result
```

```python
# Sort by Max Price
result.sort_values(by='Max_Price', ascending=True)
```

### Min Value of Tops & BottomWear

## Qn 2: Run a Python program to do run length encoding example “abbcdd --> a1b3c1d2“


Here's a Python program to perform run-length encoding (RLE) on a given string

```python
def run_length_encode(inp_str):
  encoded = ''
  count = 1

  for i in range(1, len(inp_str)):
    if inp_str[i] == inp_str[i-1]:
      count += 1
    else:
      encoded += inp_str[i-1] + str(count)
      count = 1

  encoded += inp_str[-1] + str(count)

  return encoded
```

```python
print(run_length_encode('abbbcdd'))
```
